```python
# -------------------------------
# Minimal ABI Example
# -------------------------------

from web3 import Web3

# Example: connect to a local node or testnet
RPC_URL = "https://sepolia.infura.io/v3/7a02079f42fb48f28500c2abf7e7637d"
w3 = Web3(Web3.HTTPProvider(RPC_URL))

if not w3.is_connected():
    raise Exception("Web3 connection failed")
print("Connected:", w3.is_connected())

# Replace with your contract address
CONTRACT_ADDRESS = "0x3026C8103e2e34E7535184c0e888B460652d4408"

# Minimal ABI: only includes the functions/events you want to interact with
CONTRACT_ABI = [
    {
        "inputs": [],
        "name": "getCount",
        "outputs": [{"internalType": "uint256", "name": "", "type": "uint256"}],
        "stateMutability": "view",
        "type": "function"
    },
    {
        "inputs": [],
        "name": "increment",
        "outputs": [],
        "stateMutability": "nonpayable",
        "type": "function"
    }
]

# Load contract
contract = w3.eth.contract(
    address=w3.to_checksum_address(CONTRACT_ADDRESS),
    abi=CONTRACT_ABI
)

# Example: call read-only function
count = contract.functions.getCount().call()
print(f"Current count in contract: {count}")
```
